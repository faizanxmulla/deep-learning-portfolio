##### **Defining Functions.**

```python
# Sigmoid function 

def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
# Loss function

def loss(y, y_hat):
    return 0.5 * np.sum((y - y_hat) ** 2)
```
