**Importing Libraries.**

```python
import numpy as np
```

**Given datapoints.**

```python
X = [-2, 2]
Y = [0.18, 0.81]
```

**Logistic Sigmoid function.**

<div align="center">

$\sigma(x) = \frac{1}{1 + e^{-(wx+b)}}$
</div>

```python
def sigmoid(x, w, b):
    return 1.0 / (1.0 + np.exp(-(w * x + b)))
```

**Loss function.**

<div align="center">

$\text{Loss} = \frac{1}{2} \sum_{i=1}^N (\hat{y}_i - y_i )^2$
</div>



```python
def loss(w, b):
    err = 0.0
    for x, y in zip(X, Y):
        fx = sigmoid(x, w, b)
        err += (fx - y) ** 2
    return 0.5 * err
```

**Gradient of error function w.r.t. "w"**

```python
def grad_w(w, b, x, y):
    fx = sigmoid(x, w, b)
    return (fx - y) * fx * (1 - fx) * x
```

**Gradient of error function w.r.t. "b"**

```python
def grad_b(w, b, x, y):
    fx = sigmoid(x, w, b)
    return (fx - y) * fx * (1 - fx)
```

##### **Gradient Descent Algorithm.**

```python
def gradient_descent():
    w, b, eta, max_epochs = 0, 0, 0.1, 70

    for i in range(max_epochs):
        dw, db = 0, 0
        
        for x, y in zip(X, Y):
            dw += grad_w(w, b, x, y)
            db += grad_b(w, b, x, y)

        w = w - eta * dw
        b = b - eta * db

        print(f"epoch {i}, error {loss(w, b)}, w {w}, b {b}")

    print(f"\n\nw = {w}, b = {b}")
```

```python
gradient_descent()
```
